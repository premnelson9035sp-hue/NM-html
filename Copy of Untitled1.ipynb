```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_name = "ibm-granite/granite-3.3-2b-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)

print("Model Loaded Successfully")
```

```python
class ColorPaletteGenerator:

    def __init__(self):
        self.palettes = {
            "modern": {
                "primary": "#4F46E5",
                "secondary": "#6366F1",
                "accent": "#22C55E",
                "background": "#0F172A",
                "text": "#F1F5F9"
            },

            "ocean": {
                "primary": "#0284C7",
                "secondary": "#06B6D4",
                "accent": "#38BDF8",
                "background": "#082F49",
                "text": "#E0F2FE"
            },

            "forest": {
                "primary": "#15803D",
                "secondary": "#22C55E",
                "accent": "#84CC16",
                "background": "#052E16",
                "text": "#ECFDF5"
            },

            "sunset": {
                "primary": "#EA580C",
                "secondary": "#F97316",
                "accent": "#FB923C",
                "background": "#431407",
                "text": "#FFF7ED"
            }
        }

    def get_palette(self, style):
        return self.palettes.get(style, self.palettes["modern"])
```

```python
class HTMLPageBuilder:

    def create_page(self, title, description, palette):

        html = f"""
        <!DOCTYPE html>
        <html>
        <head>
        <title>{title}</title>

        <style>
        body {{
            background:{palette["background"]};
            color:{palette["text"]};
            font-family:Arial;
            margin:0;
        }}

        header {{
            background:{palette["primary"]};
            padding:40px;
            text-align:center;
        }}

        section {{
            padding:40px;
        }}

        footer {{
            background:{palette["secondary"]};
            text-align:center;
            padding:20px;
        }}
        </style>

        </head>

        <body>

        <header>
        <h1>{title}</h1>
        </header>

        <section>
        <p>{description}</p>
        </section>

        <footer>
        <p>Generated by HTML Quick Styler</p>
        </footer>

        </body>
        </html>
        """

        return html
```

```python
palette_gen = ColorPaletteGenerator()
page_builder = HTMLPageBuilder()

def generate_page(title, description, style):

    palette = palette_gen.get_palette(style)

    html = page_builder.create_page(title, description, palette)

    return html
```

```python
import gradio as gr

with gr.Blocks() as demo:

    gr.Markdown("# HTML Quick Styler")

    title = gr.Textbox(label="Website Title")

    description = gr.Textbox(label="Website Description")

    style = gr.Dropdown(
        ["modern","ocean","forest","sunset"],
        label="Color Style"
    )

    output = gr.HTML()

    btn = gr.Button("Generate Website")

    btn.click(
        generate_page,
        inputs=[title, description, style],
        outputs=output
    )

demo.launch(share=True)
```
